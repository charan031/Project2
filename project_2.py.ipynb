```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
print("Datesheet of Matches")
Data_1 = pd.read_csv("Matches.csv")
Data_1
```

```python
print("Datasheet of Deliveries")
Data_2 = pd.read_csv("deliveries.csv")
Data_2
```

```python
print("Name of the teams:")
Data_2['batting_team'].unique()
```

```python
print("No.of Super overs:")
Data_2['is_super_over'].value_counts()
```

```python
Season_data=Data_1[['id','season','winner']]
print("Complete Data after merging few columes of Matches_Data with Deliveries_Data which we required:")
Complete_data=Data_2.merge(Season_data,how='inner',left_on='match_id',right_on='id')
Complete_data
```

```python
print("Columns of Matches_Data:")
Data_1.columns.values
```

```python
Data_1 = Data_1.drop(columns=["umpire3"],axis=1)
print("Matches_Data after droping Umpire3 column:")
Data_1.head()
```

```python
Countwins_per_season = Data_1.groupby("season")["winner"].value_counts()
print("No. of wins per season:")
Countwins_per_season
```

```python
plt.figure(figsize = (8,5))
sns.countplot(x='winner',data=Data_1,  color="Blue")
plt.title("Plot of Numbers of matches won by team ",fontsize=30)
plt.xticks(rotation=90)
plt.xlabel("Teams",fontsize=20),
plt.ylabel("No of wins",fontsize=20)
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.countplot('season',data=Data_1,palette="hsv")
plt.title("Plot of Number of Matches played in each IPL season",fontsize=30)
plt.xlabel("Season",fontsize=20)
plt.ylabel('Matches',fontsize=20)
plt.show()
```

```python
print("Pie chart Plot on Toss Result")
Toss=Data_1.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
colors = ['Green', 'red']
plt.figure(figsize = (12,10))
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Toss result',fontsize=30)
plt.show()
```

```python
plt.figure(figsize = (15,5))
sns.countplot('season',hue='toss_decision',data=Data_1,palette='winter')
plt.title("Plot of Numbers of matches won by Toss result ",fontsize=30)
plt.xlabel("Season",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.show()
```

```python
print("Pie chart Plot on Match Result on the basis of Bat_first or Bowl_first")
Data_1['win_by']=np.where(Data_1['win_by_runs']>0,'Bat first','Bowl first')
Win=Data_1.win_by.value_counts()
labels=np.array(Win.index)
sizes = Win.values
colors = ['green', 'red']
plt.figure(figsize = (12,10))
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Match Result',fontsize=30)
plt.show()
```

```python
plt.figure(figsize = (15,8))
sns.countplot('season',hue='win_by',data=Data_1,palette='cool')
plt.title("Plot of Numbers of matches won by batting and bowling first ",fontsize=30)
plt.xlabel("Season",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.show()
```

```python
print("Winners of Matches Season wise:")
Final_matches=Data_1.drop_duplicates(subset=['season'], keep='last')
Final_matches[['season','winner']].reset_index(drop=True).sort_values('season')
```

```python
print("Pie chart Plot on Winning percentage in final on the basis of what team chooses Bat/Field")
match = Final_matches.win_by.value_counts()
labels=np.array(Toss.index)
sizes = match.values
colors = ['green', 'red']
plt.figure(figsize = (12,10))
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Match Result',fontsize=30)
plt.show()
```

```python
print("Pie chart Plot on Winning percentage in final on the basis toss result")
Toss=Final_matches.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
colors = ['green', 'red']
plt.figure(figsize = (12,10))
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Toss Result',fontsize=30)
plt.show()
```

```python
print("Toss_winner, Toss_decision, winner in final matches:")
Final_matches[['toss_winner','toss_decision','winner']].reset_index(drop=True)
```

```python
print("Number of season won by teams:")
Final_matches["winner"].value_counts()
```

```python
print("IPL Finals venues and winners along with the number of wins:")
Final_matches.groupby(['city','winner']).size()
```

```python
print("Name of top players in IPL:")
plt.figure(figsize = (18,10))
top_players = Data_1.player_of_match.value_counts()[:10]
fig, ax = plt.subplots()
ax.set_ylim([0,25])
ax.set_ylabel("Count")
ax.set_title("Top player of the match Winners")
top_players.plot.bar()
sns.barplot(x = top_players.index, y = top_players, orient='v', color="blue");
plt.show()
```

```python
print("Man of the match of winning team:")
Final_matches[['winner','player_of_match']].reset_index(drop=True)
```

```python
print("No. of matches won after wining the toss in Final_matches:")
len(Final_matches[Final_matches['toss_winner']==Final_matches['winner']]['winner'])
```

```python
print("Numbers of fours hit by team:")
No_of_four=Complete_data[Complete_data['batsman_runs']==4]
No_of_four.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
```

```python
ax=No_of_four.groupby('season')['batsman_runs'].agg([('four','count')]).reset_index().plot('season','four',kind='bar',color = 'green')
plt.title("Plot of Numbers of fours hit in each season ",fontsize=30)
plt.xticks(rotation=90)
plt.xlabel("season",fontsize=20)
plt.ylabel("No of fours",fontsize=20)
plt.show()
```

```python
Four_of_batsman=No_of_four.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
ax=Four_of_batsman.iloc[:10,:].plot('batsman','four',kind='bar',color='green')
plt.title("Plot of Numbers of fours hit by players ",fontsize=30)
plt.xticks(rotation=90)
plt.xlabel("Player name",fontsize=20)
plt.ylabel("No of fours",fontsize=20)
plt.show()
```

```python
print("Numbers of sixes hit by team:")
No_of_six=Complete_data[Complete_data['batsman_runs']==6]
No_of_six.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')])
```

```python
ax=No_of_six.groupby('season')['batsman_runs'].agg([('six','count')]).reset_index().plot('season','six',kind='bar',color = 'blue')
plt.title("Plot of Numbers of Sixes hit in each season ",fontsize=30)
plt.xticks(rotation=90)
plt.xlabel("season",fontsize=20)
plt.ylabel("No of fours",fontsize=20)
plt.show()
```

```python
Six_of_batsman=No_of_six.groupby('batsman')['batsman_runs'].agg([('six','count')]).reset_index().sort_values('six',ascending=0)
ax=Six_of_batsman.iloc[:10,:].plot('batsman','six',kind='bar',color='Green')
plt.title("Numbers of six hit by players ",fontsize=30)
plt.xticks(rotation=90)
plt.xlabel("Player name",fontsize=20)
plt.ylabel("No of six",fontsize=20)
plt.show()
```

```python
print("No. of matches played by batsman are:")
Player_matches= Data_2[["match_id","player_dismissed"]]
Player_matches=Player_matches .groupby("player_dismissed")["match_id"].count().reset_index().sort_values(by="match_id",ascending=False).reset_index(drop=True)
Player_matches.columns=["batsman","No_of Matches"]
Player_matches
```

```python
print("Top 5 highest no. of matches played by players are:")
Player_matches.head(5)
```

```python
print("Least no. of matches played by players are:")
Player_matches.tail(5)
```

```python
Score_of_batsman=Data_2.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum',ascending=False).reset_index(drop=True)
Score_of_batsman=Score_of_batsman.rename(columns={'sum':'batsman_runs'})
print("Top 10 Leading Run Scorer in IPL:")
Score_of_batsman.iloc[:10,:]
```

```python
plt.figure(figsize=(8,5))
ax=sns.countplot(Data_2.dismissal_kind, color="red")
plt.title("Dismissals in IPL",fontsize=30)
plt.xlabel("Dismissals kind",fontsize=20)
plt.ylabel("count",fontsize=20)
plt.xticks(rotation=50)
plt.show()
```

```python
No_of_wicket=Data_2.dropna(subset=['dismissal_kind'])
No_of_wicket=No_of_wicket[~No_of_wicket['dismissal_kind'].isin(['run out','retired hurt','obstructing the field'])]
print("Top wicket taking bowlers in IPL are:")
No_of_wicket.groupby('bowler')['dismissal_kind'].agg(['count']).reset_index().sort_values('count',ascending=False).reset_index(drop=True).iloc[:10,:]
```
